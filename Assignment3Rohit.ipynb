**QUESTION 1**

Generate 10000 values between 5 and 500. Find its mean, median,
variance, 25th percentile.

```python
#Assigning variable 'a' to 10000 random integer values between 5 and 500
a = np.random.randint(5,500,10000)
print(a)

#Mean of all the values in array 'a'
print(np.mean(a))

#Varience of all the values in array 'a'
print(np.var(a))

# 25th percentile of the values in array 'a'
print(np.percentile(a,25))
```

Create a new column called “Bool” which has value as “yes” if
Survived=0 and “no” if Survived=1.

```python
# Using the 'where' function to check for a condition and accordingly create a new column called 'Bool' storing either 'True' or 'False'

data['Bool'] = np.where(data['Survived']== 1, True, False)
data
```
